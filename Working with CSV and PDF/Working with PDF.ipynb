```python
x = open('Working_Business_Proposal.pdf','rb')
pdf_read = PyPDF2.PdfFileReader(x)
```

```python
new_page = pdf_read.getPage(3)
```

```python
pdf_write = PyPDF2.PdfFileWriter()
```

```python
pdf_write.addPage(new_page)
```

```python
new_file = open('new.pdf','wb')
```

```python
pdf_write.write(new_file)
```

```python
x.close()
new_file.close()
```
