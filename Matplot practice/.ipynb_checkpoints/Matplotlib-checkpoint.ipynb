```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Simple Graph
x = np.arange(0,10,2)

plt.title('Basic Graph',fontdict={'fontsize':15, 'fontname':'Comic Sans MS'})

plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.plot(x,np.sin(x),'b.--',label='Sin x', markersize=10,markeredgecolor = 'r',linewidth=3)
plt .plot(x,x*2,'r^-',label = "X^2")

plt.legend()

plt.savefig('first_graph.png',dpi=300)
plt.show()
```

```python
#Bar Graph
x=[13,55,30,59,11]
y=['A','B','C','D','E']
colors=['r','b','cyan','#abcdef','#0e0e0e']
plt.bar(y,x,color=colors,label=y)
plt.legend()
plt.show()
```

```python
#Gas price graph
gas = pd.read_csv('gas_prices.csv')
```

```python
gas
```

```python
plt.figure(figsize = (8,8))
plt.title('Gas prices in USD',fontdict={'fontsize':20,'fontname':'Arial'})

countries = ['USA','Mexico','France','Germany']
for country in countries:
    plt.plot(gas.Year,gas[country],marker='.',markersize=10, label=country)
    
plt.xticks(gas.Year[::3])
plt.xlabel('Year',fontdict={'fontsize':15,'fontname':'Arial'})
plt.ylabel('Price',fontdict={'fontsize':15,'fontname':'Arial'})
    
plt.legend()
plt.show()
plt.savefig('Gas_Price.png',dpi=250)
```

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa
```

```python
fifa.Wage[2]
```

```python
fifa.Wage= [x.strip('€') for x in fifa.Wage]
fifa.Wage= [int(x.strip('K')) for x in fifa.Wage]
```

```python
fifa.Wage
```

```python
low = fifa.loc[fifa.Wage<100].count()[0]
below_average = fifa.loc[(fifa.Wage>=100) & (fifa.Wage<250)].count()[0]
avearge = fifa.loc[(fifa.Wage>=250) & (fifa.Wage<350)].count()[0]
above_average = fifa.loc[(fifa.Wage>=350) & (fifa.Wage<400)].count()[0]
high = fifa.loc[fifa.Wage>=400].count()[0]

plt.figure(figsize=(5,5))
label = ['under 100','100-250','250-350','350-400','above 400']
wages = [low,below_average,avearge,above_average,high]
explode = [1,1,1,1,1]
plt.pie(wages,autopct='%.2f %%',pctdistance = 0.8,explode = explode,labels=label)
plt.show()
```
