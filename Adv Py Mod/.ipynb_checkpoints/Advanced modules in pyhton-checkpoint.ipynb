# Regular Expressions

```python
import re
```

```python
text = "I am Pranav learning python, this is test for re, this is part one"
```

```python
pattern = 'this'
```

```python
re.search(pattern,text)
```

```python
re.findall(pattern,text)
```

```python
for i in re.finditer(pattern,text):
    print(i.span())
    print(i)
```

```python
p = "The number is 372-946-3922"
```

```python
num = re.search(r'\d\d\d-\d\d\d-\d\d\d\d',p)
```

```python
num
```

```python
num = re.search(r'\d{3}-\d{3}-\d{4}',p)
```

```python
num
```

```python
te = re.search(r'\w{3,}',p)
```

```python
te
```

```python
te.end()
```

```python
sep = re.compile(r'(\d{3})-(\d{3})-(\d{4})')
```

```python
g = re.search(sep,p)
```

```python
g
```

```python
g.group(2)
```

```python
#OR operator
re.search(r'me|I',"I am a boy")
```

```python
#Wildcard operator
re.findall(r'.at','The cat in the hat sat here.')
```

```python
#startwith
re.search(r'^\d+',"22 is the num")
```

```python
#endwith
re.findall(r'\d$',"u are 2")
```

```python
#Exclusion operator
re.findall(r'[^\d]+',"you are 28 and she is 29")
```

# Timing your code

```python
def func_1(n):
    for i in range(1,n+1):
        print(i*3)
```

```python
def func_2(n):
    result = [i*3 for i in range (1,n+1)]
    print(result)
```

```python
#time elapsed method
import time
```

```python
t1 = time.time()
func_1(10)
t2 =time.time()
print(t2- t1)
```

```python
t1 = time.time()
func_2(10)
t2 =time.time()
print(t2- t1)
```

```python
#timeit module
import timeit
```

```python
stmt = '''
func_1(10)
'''
setup = '''
def func_1(n):
    for i in range(1,n+1):
        print(i*3)
'''
```

```python
timeit.timeit(stmt,setup,number = 500)
```

```python
stmt = '''
func_2(10)
'''
setup = '''
def func_2(n):
    result = [i*3 for i in range (1,n+1)]
    print(result)
'''
```

```python
timeit.timeit(stmt,setup,number = 500)
```

```python
%%timeit
func_1(2)
```
